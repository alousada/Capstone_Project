The data related to the problem will be collected from an external site* using scraping techniques with the Beatiful Soup library on Python.

*the chosen site is https://www.guiamais.com.br/sao-paulo-sp - onde site where it is possible to identify the possible locations of our problem with addresses and postal codes.

The features extracted will be:

- Store Name
- Locality (city)
- Postal Code
- Street Address
- Number
- Neighborhood

The data is collected direct from site using the Pandas library with the command read_html. All non related data to our key objective found brewery locations should be filtered and excluded from our analysis.

The geolocation data will be a added feature based on the postal code and addresses found using the Nominatim library.
